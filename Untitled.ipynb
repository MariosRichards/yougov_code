```python
# Columns
Leadership_Vote_2015 = [ "Jeremy Corbyn", "Yvette Cooper", "Andy Burnham", "Liz Kendall", "Did not vote" ]
Membership_Length    = [ "Pre 2015 GE", "During 2015 Leadership Election", "After JC Elected" ]
EU_Ref_Vote          = [ "Remain", "Leave" ]
GE_Vote_2015         = [ "Labour", "Not Labour" ]
Gender               = [ "Male", "Female" ]
Age                  = [ "18-24", "25-39", "40-59", "60+" ]
Social_Grade         = [ "ABC1", "C2DE" ]
Region               = [ "London", "Rest of South", "Midlands / Wales",	"North", "Scotland" ]

yougov_categories = [Leadership_Vote_2015, Membership_Length, EU_Ref_Vote, GE_Vote_2015, Gender, Age, Social_Grade, Region]
```

```python
# (0) Reconstruction

ft.loc[ ft.index[1]  ] # unweighted sample
ft.loc[ ft.index[2]  ] /100 # have voted
ft.loc[ ft.index[4:] ] /100 # Owen Smith/Jeremy Corbyn/Dont know
# maybe add this as a row?
ft.loc[ ft.index[1]  ] * (ft.loc[ ft.index[3]  ] /100) # did not vote in LE

raw =  ft.loc[ ft.index[1]  ] * (ft.loc[ ft.index[2]  ] /100) * (ft.loc[ ft.index[4:] ] /100)

# raw.sum(axis=0)
# raw
for cat in yougov_categories:
    raw[cat] = raw[cat]*986/raw[cat].sum().sum()
    print( raw[cat].sum().sum() )

#print( raw[yougov_categories[1]].sum().sum() )

raw[cat].sum().sum()

# function takes a (part of a) DataFrame
# which has real numbers, but which sums to an integer
# finds the (integer) difference between that sum and the sum of the integer parts of all values
# assign those individuals
# deterministic option - assign by descending order of remainder size (can't have more assignees than remainders over 0)
# stochastic option    - assign by chance depending on proportion of all fractions

for cat in yougov_categories:

    int_raw = np.floor(raw[cat]).astype('int') # rounded down version of data frame
    to_be_assigned = 986 - int_raw.sum().sum()

    remainders = raw[cat] - int_raw # real remainders of array
    thresh = remainders.stack().nlargest(to_be_assigned).min() # 
    c = int_raw[cat].where( remainders < thresh, int_raw[cat]+1 )
    raw[cat] = c
    print(cat)
    print(raw[cat].sum().sum())
    # remainders.values.flatten().sort()
    # sort on values, return indices, take top X indices by size, use to increment int_raw
```

```python
# (0) Reconstruction
# using weighted samples instead
ft.loc[ ft.index[0]  ] # weighted sample
ft.loc[ ft.index[2]  ] /100 # have voted
ft.loc[ ft.index[4:] ] /100 # Owen Smith/Jeremy Corbyn/Dont know
# maybe add this as a row?
ft.loc[ ft.index[0]  ] * (ft.loc[ ft.index[3]  ] /100) # did not vote in LE

raw =  ft.loc[ ft.index[0]  ] * (ft.loc[ ft.index[2]  ] /100) * (ft.loc[ ft.index[4:] ] /100)

# raw.sum(axis=0)
# raw
for cat in yougov_categories:
    raw[cat] = raw[cat]*986/raw[cat].sum().sum()
    print( raw[cat].sum().sum() )

#print( raw[yougov_categories[1]].sum().sum() )

raw[cat].sum().sum()

# function takes a (part of a) DataFrame
# which has real numbers, but which sums to an integer
# finds the (integer) difference between that sum and the sum of the integer parts of all values
# assign those individuals
# deterministic option - assign by descending order of remainder size (can't have more assignees than remainders over 0)
# stochastic option    - assign by chance depending on proportion of all fractions

for cat in yougov_categories:

    int_raw = np.floor(raw[cat]).astype('int') # rounded down version of data frame
    to_be_assigned = 986 - int_raw.sum().sum()

    remainders = raw[cat] - int_raw # real remainders of array
    thresh = remainders.stack().nlargest(to_be_assigned).min() # 
    c = int_raw[cat].where( remainders < thresh, int_raw[cat]+1 )
    raw[cat] = c
    print(cat)
    print(raw[cat].sum().sum())
    # remainders.values.flatten().sort()
    # sort on values, return indices, take top X indices by size, use to increment int_raw
```

```python
from scipy.stats import chi2_contingency
# chi2 : float
# The test statistic.
# p : float
# The p-value of the test
# dof : int
# Degrees of freedom
# expected : ndarray, same shape as observed
# The expected frequencies, based on the marginal sums of the table.
for cat in yougov_categories:
    obs = raw[cat].values
    [chi2, p, dof, expected] = chi2_contingency(obs)
    print(cat, p, chi2, dof)
```
